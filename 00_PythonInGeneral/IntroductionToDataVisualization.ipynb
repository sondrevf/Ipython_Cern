## 1.1 - Plotting multiple graphs  

```python
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt 
```

```python
# Multiple curves on the same axis
t= np.linspace(0,10,100)
temperature = 10+10*np.sin(t)
dewpoint = 50+50*np.cos(t)

plt.plot(t,temperature,'r')
plt.plot(t,dewpoint,'b')
plt.xlabel('Date')
plt.title('Temperature & Dew point')
plt.show()
```

```python
# Multiple axes
plt.figure()
plt.axes([0.05,0.05,0.425,0.9])  # [left,bottom,width, height]
plt.plot(t,temperature,'r')
plt.xlabel('Date')
plt.title('Temperature')

plt.axes([0.525,0.05,0.425,0.9])
plt.plot(t,dewpoint,'b')
plt.xlabel('Date')
plt.title('Dew point')
plt.show()
```

```python
plt.subplot(2,1,1)  #nrow,ncol,nsubplot
plt.plot(t,temperature,'r')
plt.xlabel('Date')
plt.title('Temperature')

plt.subplot(2,1,2)  #nrow,ncol,nsubplot
plt.plot(t,dewpoint,'b')
plt.xlabel('Date')
plt.title('Dew point')
plt.tight_layout()
plt.show()
```

## 1.2 - Customizing axes

```python
yr = np.arange(1940,2010)
gdp = np.exp((yr-1940)/20)*1e3

plt.plot(yr,gdp)
plt.xlabel('Year') ; plt.ylabel('Billions of Dollars') ; plt.title('US Gross Domestic Product')
plt.show()

plt.plot(yr,gdp)
plt.xlabel('Year') ; plt.ylabel('Billions of Dollars') ; plt.title('US Gross Domestic Product')
plt.xlim(1947,1957) ; plt.ylim(1000,2500)
plt.axis([1947,1957,1000,2500])
plt.show()
```

```python
R = np.ones(100)
phi = np.linspace(0,2*np.pi,100)
x = R*np.cos(phi)
y = R*np.sin(phi)

for i in range(4):
    plt.subplot(2,2,1+i)
    plt.plot(x,y)
    name = ['off','equal','square','tight'][i]
    plt.axis(name)
    plt.title(name)


plt.tight_layout()
plt.show()
```

## 1.3 - Legends, annotations and styles

```python
print(plt.style.available)
# plt.style.library
# plt.style.use('')
```

```python
N = 20
for i in range(3):
    length = np.random.uniform(3.5,5.5,N) - i
    width  = np.random.uniform(0.5,2.5,N) + i
    plt.scatter(length,width,marker='o',color = ['r','g','b'][i],label=['setosa','versicolor','virginica'][i])
plt.legend(loc=0)
plt.title('Iris data') ; plt.xlabel('sepal length (cm)') ; plt.ylabel('sepal width (cm)')

plt.annotate('setosa',xy= (4,1))
plt.annotate('versicolor',xy=(3,2))
plt.annotate('virginica',xy=(2,3),xytext=(1.5,1.5),arrowprops={'color':'blue','width':1})

plt.show()
```
