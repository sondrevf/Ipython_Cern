# Waterfall with varying x-axes

```python
nt = 10
nx = 100
x0 = np.linspace(-1,1,nx)*np.pi
y = np.sin(x0)

base = x0
AllY = np.zeros((nt,np.size(x0)))
for i in range(nt):
    x = x0+np.random.uniform(-.1,.1,1)
    
    f = sc.interpolate.interp1d(x,y,kind='linear',bounds_error=False,fill_value=0)
    AllY[i] = f(base)
AllY[0,0]=3
plt.figure()
plt.imshow(np.flipud(AllY),aspect='auto',extent=(-1,1,0,10),interpolation=['bilinear','none'][0])
plt.colorbar()

plt.figure()
levels = np.linspace(np.min(AllY),np.max(AllY),100)
plt.contourf(AllY,levels=levels,extent=(-1,1,0,10))
plt.colorbar()
```
