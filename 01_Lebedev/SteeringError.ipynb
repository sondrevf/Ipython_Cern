```python
def emit(x,p):
    mx = np.mean(x)
    mp = np.mean(p)
    emx = np.mean( (x-mx)**2 + (p-mp)**2 )*0.5
    return mx,mp,emx

N = int(1e6)
Nsteps=50
Dmu = 0.01*2*np.pi
cov = np.eye(2)
sep = 1
for j,dx in enumerate([1]):
    mean = [dx,0]
    x, p = np.random.multivariate_normal(mean, cov, N).T
    phi = np.random.normal(0,sep,N)*Dmu
    COS = np.cos(phi)
    SIN = np.sin(phi)

    # plt.figure()
    # plt.hist(phi,bins = int(0.5*round(np.sqrt(N))))
    # plt.figure()
    # plt.hist2d(x,p,bins = int(0.5*round(np.sqrt(N))))


    emits = np.zeros(Nsteps)
    COM = np.zeros(Nsteps)

    for i in range(Nsteps):
        if i >0:
            temp = x
            x = COS*x    + SIN*p
            p = -SIN*temp+ COS*p
        out = emit(x,p)
        emits[i] = out[2]
        COM[i] = np.sqrt(out[0]**2+out[1]**2)


    fig = plt.figure(figsize = (15,5))
    ax1 = plt.subplot(121)
    ax1.hist2d(x,p,bins = 50) ;
    ax2 = plt.subplot(122)
    ax2.hist(x,bins = 50,normed=True)
    ax2.plot(np.linspace(-5,5,100),np.exp(-np.linspace(-5,5,100)**2/2)/np.sqrt(2*np.pi))
    # plt.close('all')
    
    
    if j == 0:
        fig = plt.figure(1)
        ax1 = plt.subplots()[1]
        ax2 = ax1.twinx()
    
    ax1.tick_params('y', colors='b')
    ax1.set_ylim(1,1+0.5*sep**2*1.1)
    ax1.plot(np.arange(Nsteps)*Dmu/(2*np.pi),emits,'b')
    ax1.set_ylabel(r'Emittance $[\epsilon_0]$',color= 'b')
    ax1.set_xlabel(r'$\Delta Q_{1\sigma} \cdot $ turns')
    
    ax2.tick_params('y', colors='r')
    ax2.set_ylim(0,sep*1.1)
    ax2.plot(np.arange(Nsteps)*Dmu/(2*np.pi),COM,'r')
    ax2.set_ylabel(r'COM [$\sigma_x$]',color= 'r')
    plt.tight_layout()
    plt.savefig('COMvsEmittance_filamentation.eps')

```
